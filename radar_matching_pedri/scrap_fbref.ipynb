```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
import time
import numpy as np

from webdriver_manager.chrome import ChromeDriverManager
```

```python
options = webdriver.ChromeOptions()
options.set_capability(
    "goog:loggingPrefs", {"performance": "ALL", "browser": "ALL"}
)


driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()), options=options)
```

```python
driver.set_page_load_timeout(10)

try:
    driver.get("https://fbref.com/en/comps/12/stats/La-Liga-Stats")
except:
    pass
```

```python
time.sleep(5)

html = driver.page_source
clean_html = html.replace('', '')
df_list = pd.read_html(clean_html, attrs={'id': 'stats_standard'})
df = df_list[0]
```

```python
df
```

```python
df.columns = df.columns.droplevel(0)
df = df[df['Player'] != 'Player'].copy()
```

```python
df.head()
```

```python
df.shape
```

```python
df_liga = df[df['Pos'].str.contains('MF', na=False)].copy()
```

```python
df_liga = df_liga.reset_index(drop=True)
```

```python
df_liga.head()
```

```python
df_liga
```

```python
df_liga = df_liga.drop(['Rk'],axis=1)
```

```python
df_liga
```

```python
df_understat = pd.read_csv('league-players.csv', sep=';')
```

```python
df_understat
```

```python
df_understat = df_understat.drop(['number'],axis=1)
```

```python
df = pd.merge(df_liga, df_understat, on='Player', how='left')
```

```python
df
```

```python
df.columns
```

```python
df = df.drop(columns=['Matches'])
```

```python
df.head()
```

```python
df = df.drop(columns=['PrgR','G+A-PK','xG+xAG'])
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns=['Nation', 'Pos', 'Squad', 'Age', 'Born',
                      'MP','Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt', 'CrdY',
                       'CrdR', 'xG', 'npxG', 'npxG+xAG','Starts'])
```

```python
df.head()
```

```python
df = df.drop(df.columns[3], axis=1)
```

```python
df.head()
```

```python
df = df.fillna(value='0')
```

```python
df.dtypes
```

```python
df = df.astype({
    'Min': int, 
    '90s': float,
    'PrgC': float,
    'PrgP' : float,
    'xGChain90' : float,
    'xGBuildup90' : float
})
```

```python
df['PrgC90'] = (df['PrgC'] / df['90s']).replace([np.inf, -np.inf], 0)
df['PrgP90'] = (df['PrgP'] / df['90s']).replace([np.inf, -np.inf], 0)
```

```python
df.head()
```

```python
df = df.drop(columns=['PrgC', 'PrgP'])
```

```python
df.to_csv("fbref.csv")
```
